```python
# SIMPLE CONDITION PREDICTION FUNCTION
# Copy-paste this into your existing code with trained Logistic Regression model

import re
import numpy as np
from scipy.sparse import hstack
from sklearn.preprocessing import LabelEncoder # Import LabelEncoder

def predict_condition(review_text, condition_model, tfidf_vectorizer, scaler, label_encoder, rating=7, useful_count=50):
    """
    Predict medical condition using Logistic Regression

    Args:
        review_text (str): Patient review text
        condition_model: Trained model for condition prediction
        tfidf_vectorizer: Fitted TF-IDF vectorizer
        scaler: Fitted StandardScaler
        label_encoder: Fitted LabelEncoder for conditions
        rating (int): Rating 1-10 (default=7)
        useful_count (int): Useful count (default=50)

    Returns:
        str: Predicted medical condition
    """
    # Clean text
    text = str(review_text).lower()
    text = re.sub(r'[^\w\s]', '', text)
    text = ' '.join(text.split())

    # Convert to features
    text_features = tfidf_vectorizer.transform([text])
    numeric_features = scaler.transform([[rating, useful_count]])
    combined_features = hstack([text_features, numeric_features])

    # Predict condition using the provided model
    condition_pred_encoded = condition_model.predict(combined_features)[0]

    # Decode the predicted label back to its original string representation
    predicted_condition = label_encoder.inverse_transform([condition_pred_encoded])[0]

    return predicted_condition
```

```python
import pandas as pd
import numpy as np
import re
import joblib
from scipy.sparse import hstack

# NLTK and Sklearn imports
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
import nltk
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression

# --- 1. DATA LOADING AND CLEANING (from your file) ---
print("--- Stage 1: Loading and Cleaning Data ---")
df_train = pd.read_csv("/content/drive/MyDrive/datasets/drugscom_train.csv")
df_test = pd.read_csv("/content/drive/MyDrive/datasets/drugscom_test.csv")
df = pd.concat([df_train, df_test], ignore_index=True)

mask = df['condition'].str.split().str.len() <= 4
df_filtered = df[mask]
top_40_conditions = df_filtered['condition'].value_counts().head(40).index
df_modified = df_filtered.copy()
df_modified['condition'] = np.where(df_modified['condition'].isin(top_40_conditions), df_modified['condition'], 'Other')

other_indices = df_modified[df_modified['condition'] == 'Other'].index
if len(other_indices) >= 35000:
    indices_to_drop = other_indices[:35000]
    df_final = df_modified.drop(indices_to_drop)
else:
    df_final = df_modified.copy()

df_final.drop(columns=['Unnamed: 0', 'date'], axis=1, inplace=True, errors='ignore')
print("Data loading and cleaning complete.")

# --- 2. DATA SPLITTING (from your file) ---
print("\n--- Stage 2: Splitting Data ---")
X = df_final[['review', 'rating', 'usefulCount']]
y_condition = df_final['condition']
X_train, X_test, y_condition_train, y_condition_test = train_test_split(
    X, y_condition, test_size=0.3, random_state=42, stratify=y_condition
)
print("Data splitting complete.")

# --- 3. TEXT PREPROCESSING (from your file) ---
print("\n--- Stage 3: Preprocessing Text ---")
nltk.download('punkt', quiet=True)
nltk.download('stopwords', quiet=True)
# Download punkt_tab resource
nltk.download('punkt_tab', quiet=True)

stop_words = set(stopwords.words('english'))
stemmer = PorterStemmer()

def preprocess_text(text):
    text = re.sub(r'[^\w\s]', '', str(text).lower())
    tokens = word_tokenize(text)
    processed_tokens = [stemmer.stem(word) for word in tokens if word not in stop_words]
    return " ".join(processed_tokens)

X_train['processed_review'] = X_train['review'].apply(preprocess_text)
X_test['processed_review'] = X_test['review'].apply(preprocess_text)
print("Text preprocessing complete.")

# --- 4. FEATURE ENGINEERING AND LABEL ENCODING (Corrected) ---
print("\n--- Stage 4: Engineering Features ---")
# Text Features
tfidf = TfidfVectorizer(max_features=7000, ngram_range=(1, 2))
X_train_text = tfidf.fit_transform(X_train['processed_review'])
X_test_text = tfidf.transform(X_test['processed_review'])

# Numeric Features
X_train['log_usefulCount'] = np.log1p(X_train['usefulCount'])
X_test['log_usefulCount'] = np.log1p(X_test['usefulCount'])
numeric_features = ['rating', 'log_usefulCount']
scaler = StandardScaler()
X_train_numeric = scaler.fit_transform(X_train[numeric_features])
X_test_numeric = scaler.transform(X_test[numeric_features])

# Combination
X_train_final = hstack([X_train_text, X_train_numeric])
X_test_final = hstack([X_test_text, X_test_numeric])

# **NEW AND CRITICAL STEP: Label Encoding**
le_condition = LabelEncoder()
y_condition_train_encoded = le_condition.fit_transform(y_condition_train)
y_condition_test_encoded = le_condition.transform(y_condition_test)
print("Feature engineering complete.")

# --- 5. MODEL TRAINING (from your file) ---
print("\n--- Stage 5: Training Model ---")
condition_model = LogisticRegression(penalty='l2', solver='saga', random_state=42)
condition_model.fit(X_train_final, y_condition_train_encoded)
print("Model training complete.")

# --- 6. PREDICTION FUNCTION AND EXECUTION (New Part) ---
print("\n--- Stage 6: Making a Prediction ---")
def predict_condition(review_text, rating=7, useful_count=50):
    """Predicts a condition using the objects already in memory."""
    # Preprocess the input text
    processed_text = preprocess_text(review_text)

    # Transform features using the fitted objects
    text_features = tfidf.transform([processed_text])
    numeric_features = scaler.transform([[rating, np.log1p(useful_count)]])
    combined_features = hstack([text_features, numeric_features])

    # Predict the encoded label
    predicted_encoded_label = condition_model.predict(combined_features)[0]

    # Decode the label to the original condition name
    predicted_condition = le_condition.inverse_transform([predicted_encoded_label])[0]

    return predicted_condition

```

```python
# --- Let's predict a condition for a new review ---
my_review = "Before starting this medication, I often felt discomfort in my stomach and had difficulty during bathroom visits. Since beginning the treatment, those issues have eased significantly. My digestion feels smoother, and daily routines are much more comfortable. The overall improvement has made a noticeable difference in my comfort and well-being."


my_rating = 6
my_useful_count = 40

predicted_result = predict_condition(my_review, my_rating, my_useful_count)

print("\n" + "="*30)
print("      PREDICTION RESULT")
print("="*30)
print(f"Review Text: '{my_review}'")
print(f"Predicted Condition: {predicted_result}")
print("="*30)
```
